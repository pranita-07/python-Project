```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = data.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
sales_gen = data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(10, 6))
sales_gen.plot(kind='bar', color='lightcoral')
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Group by 'Category' and calculate the total sales for each category
category_sales = data.groupby('Category')['Amount'].sum().sort_values(ascending=False)
```

```python
# Plot the sales distribution by category
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Convert 'Date' column to datetime if it's not already
data['Date'] = pd.to_datetime(data['Date'])

# Extract the month and year for grouping
data['YearMonth'] = data['Date'].dt.to_period('M')
```

```python
# Group by YearMonth and calculate the total sales
monthly_sales = data.groupby('YearMonth')['Amount'].sum()

# Plot the sales over time
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Total Sales Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Group by 'Gender' and calculate the total sales for each gender
gender_sales = data.groupby('Gender')['Amount'].sum()

# Plot the sales distribution by gender
plt.figure(figsize=(8, 8))
gender_sales.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Sales Distribution by Gender')
plt.ylabel('')
plt.show()
```

```python
# Group by 'Size' and calculate the total sales for each size
size_sales = data.groupby('Size')['Amount'].sum().sort_values(ascending=False)

# Plot the sales distribution by size
plt.figure(figsize=(10, 6))
size_sales.plot(kind='bar', color='orange')
plt.title('Total Sales by Product Size')
plt.xlabel('Product Size')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Group by 'ship-city' and calculate the total sales for each city
city_sales = data.groupby('ship-city')['Amount'].sum().sort_values(ascending=False)

# Plot the sales distribution by city
plt.figure(figsize=(12, 8))
city_sales.head(20).plot(kind='bar', color='green')
plt.title('Top 20 Cities by Total Sales')
plt.xlabel('City')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Calculate monthly growth rate
monthly_sales = data.groupby('YearMonth')['Amount'].sum()
monthly_growth_rate = monthly_sales.pct_change() * 100

# Plot the monthly growth rate
plt.figure(figsize=(10, 6))
monthly_growth_rate.plot(kind='bar', color='purple')
plt.title('Monthly Sales Growth Rate (%)')
plt.xlabel('Month-Year')
plt.ylabel('Growth Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Group by 'Category' and calculate the total sales
product_sales = data.groupby('Category')['Amount'].sum().sort_values(ascending=False)

# Plot the top 10 selling products
plt.figure(figsize=(10, 6))
product_sales.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Selling Products')
plt.xlabel('Product Category')
plt.ylabel('Total Sales (INR)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
